```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

```python
I = plt.imread('ny.jpg')

#I.flatten() turns an ndarray to a one dimensional ndarray
print("I has the values...")
print(str(np.unique(I.ravel()))) #https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.ravel.html
```

```python
n, allbins, patches = plt.hist(I.ravel(), bins=np.unique(I.ravel()))
plt.title("Overall Histogram")
```

```python
#https://stackoverflow.com/questions/6871201/plot-two-histograms-at-the-same-time-with-matplotlib
plt.figure()

plt.hist(I[...,0].ravel(), allbins, alpha = .6, label = 'red', color = 'r');
plt.hist(I[...,1].ravel(), allbins, alpha = .6, label = 'green', color = 'g');
plt.hist(I[...,2].ravel(), allbins, alpha = .6, label = 'blue', color = 'b');
```

```python
f, axarr = plt.subplots(1,2, figsize=(10, 3))

axarr[0].imshow(I) #https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html
axarr[0].set_title('Image')

axarr[1].hist(I[...,0].ravel(), allbins, alpha = .6, label = 'red', color = 'r');
axarr[1].hist(I[...,1].ravel(), allbins, alpha = .6, label = 'green', color = 'g');
axarr[1].hist(I[...,2].ravel(), allbins, alpha = .6, label = 'blue', color = 'b');
axarr[1].legend(loc = 'upper right');
plt.show()
```

```python
I.shape
```

```python
newI = I.copy()
redchange = 30 # red chnage
bluechange = 10 #blue and green change
for y in range(0,532): #for all pixles of the picture
    for x in range(0,798):
        if(newI[y][x][0] <= redchange): #check for wrap around of 256
            newI[y][x][0] = 0
        else: newI[y][x][0] -= redchange #makes it less red
        
        if(newI[y][x][2]+bluechange >= 255): #makes it more blue and green
            newI[y][x][2] += bluechange
        
        if(newI[y][x][1]+bluechange >= 255):
            newI[y][x][1] += bluechange


f, axarr = plt.subplots(1,2, figsize=(10, 3))

axarr[0].imshow(newI) #https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html
axarr[0].set_title('New Image')
axarr[1].imshow(I) #https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html
axarr[1].set_title('Old Image')


plt.show()
```

```python
f, axarr = plt.subplots(1,2, figsize=(10, 3))
axarr[0].hist(I[...,0].ravel(), allbins, alpha = .6, label = 'red', color = 'r');
axarr[0].hist(I[...,1].ravel(), allbins, alpha = .6, label = 'green', color = 'g');
axarr[0].hist(I[...,2].ravel(), allbins, alpha = .6, label = 'blue', color = 'b');
axarr[0].legend(loc = 'upper right');
axarr[0].set_title('Old Image')


axarr[1].hist(newI[...,0].ravel(), allbins, alpha = .6, label = 'red', color = 'r');
axarr[1].hist(newI[...,1].ravel(), allbins, alpha = .6, label = 'green', color = 'g');
axarr[1].hist(newI[...,2].ravel(), allbins, alpha = .6, label = 'blue', color = 'b');
axarr[1].axes.set_ylim([0,10000])
axarr[1].legend(loc = 'upper right');
axarr[0].set_title('New Image')
plt.show()
```
